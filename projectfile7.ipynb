```python
import glob
import os
import pandas as pd
import numpy as np

xlxs_path = '/Users/soiji/Documents/GitHub/DataProcessing/Data'
xlxs_list = sorted(os.listdir('Data'))
```

```python
data = pd.read_excel(xlxs_path + '/' + str(xlxs_list[6]), engine='openpyxl')
print(data.head())
print(data.shape)
```

```python
data.info()
```

```python
column_list = data.columns.tolist()
data.columns = [x.strip() for x in column_list]
```

```python
import re

reg = r".*[A-z]+.*"

column_list = data.columns[2:]
find_idx = []
data_list = []

## object to string, find using regex, find row(data) by index number or value and change it
for c in column_list:
    find_data = []
    for i in data[c].astype('string'):
        find = re.findall(reg, i)
        find_str = ''.join(find)
        if find_str != '':
            find_data.append(find_str)
            data_list.append(find_str)
    idx = data[data[c].isin(find_data)].index.tolist()
    find_idx.extend(idx)
```

```python
find_idx = list(set(find_idx))
```
